```
# 1. Get a list of tuples.Write a function to sort the lsit by second value.
def sort_by_second(l):
  for i in range(0,len(l)):
    for j in range(0,len(l)-1-i):
      if l[j][1]>l[j+1][1]:
        temp = l[j]
        l[j] = l[j+1]
        l[j+1] = temp
  return l
```

```
print(sort_by_second([(3,0),(0,10),(2,1),(5,3)]))
```

```
# 4 . given the coordinates of the equilateral triangle calculate the circumcenter.
#value of sin 120 = root(3)/2 == 0.866
def calc_circumcenter(x1,y1,x2,y2,x3,y3):
  x_center = ((x1*0.866)+(x2*0.866)+(x3*0.866))/3*0.866
  y_center = ((y1*0.866)+(y2*0.866)+(y3*0.866))/3*0.866
  return (x_center,y_center)
```

```
#dist between two vertices:
from math import sqrt

def dist(x1,y1,x2,y2):
  return sqrt(((x2-x1)**2 )+ ((y2-y1)**2 ))
```

```
# 5 . 

def cal_inner_vertices(x1,y1,x2,y2,x3,y3,N):
  
  triangles = []
  side_len = dist(x1,y1,x2,y2)
  dis_x = side_len/N
  midx = (x1+x3)/2
  midy = (y1+y3)/2
  v_stack = []
  y_len = dist(x2,y2,midx,midy)
  dis_y = y_len/N
  v_stack.append([x2,y2])
  while(v_stack):
    X1 = v_stack[0][0]
    Y1 = v_stack[0][1]
    X2 = X1 - dis_x
    Y2 = Y1-dis_y
    X3 = X1+dis_x
    Y3 = Y1-dis_y
    triangles.append((X1,Y1,X2,Y2,X3,Y3))
    if Y3>midy:
        v_stack.append([X2,Y2])
        v_stack.append([X3,Y3])
    v_stack.pop(0)
  return triangles
```

```
#example vertices x1,y1,x2,y2,x3,y3 are 0,0,5,8.66,10,0 respectively  and N=5.
print("Vertices of the inner triangles as follows")
print(cal_inner_vertices(0,0,5,8.66,10,0,5))
```
